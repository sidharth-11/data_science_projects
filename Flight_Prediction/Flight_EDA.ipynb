### Feature Engineering Process

```python
df_final['Day']=df_final['Date_of_Journey'].str.split('/').str[0]
df_final['Month']=df_final['Date_of_Journey'].str.split('/').str[1]
df_final['Year']=df_final['Date_of_Journey'].str.split('/').str[2]
df_final.head()
```

```python
df_final['Day']=df_final['Day'].astype('int64')
df_final['Month']=df_final['Month'].astype('int64')
df_final['Year']=df_final['Year'].astype('int64')
```

```python
df_final.info()
```

```python
#droping the date of journey feature
df_final.drop('Date_of_Journey',axis=1,inplace=True)
df_final.head()
```

```python
##making changes in arrivaal time
df_final['Arrival_Time']=df_final['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df_final.head()
```

```python
##Replacing arrival time with hour and min field
df_final['Arival_Hours']=df_final['Arrival_Time'].str.split(':').str[0]
df_final['Arival_Mins']=df_final['Arrival_Time'].str.split(':').str[1]
```

```python
##Converting the arrival_hour and arrival_min fields to integer
df_final['Arival_Hours']=df_final['Arival_Hours'].astype('int64')
df_final['Arival_Mins']=df_final['Arival_Mins'].astype('int64')
```

```python
df_final.info()
```

```python
##droping the Arrival_Time field
df_final.drop('Arrival_Time',axis=1,inplace=True)
df_final.head()
```

```python
##Replacing dep time with hour and min field
df_final['Dep_Hours']=df_final['Dep_Time'].str.split(':').str[0]
df_final['Dep_Mins']=df_final['Dep_Time'].str.split(':').str[1]
```

```python
##Converting the Dep_hour and Dep_min fields to integer
df_final['Dep_Hours']=df_final['Dep_Hours'].astype('int64')
df_final['Dep_Mins']=df_final['Dep_Mins'].astype('int64')
```

```python
##droping the Dep_Time field
df_final.drop('Dep_Time',axis=1,inplace=True)
df_final.head()
```

```python
#Checking the number of stops field
df_final['Total_Stops'].value_counts()
```

```python
df_final['Total_Stops'].mode()[0]
```

```python
df_final['Total_Stops']=df_final['Total_Stops'].fillna(df_final['Total_Stops'].mode()[0])
df_final['Total_Stops'].unique()
```

```python
df_final['Total_Stops']=df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df_final.head()
```

```python
#Droping the route column
df_final.drop('Route',axis=1,inplace=True)
df_final.head()
```

```python
#Checking Additiion_info column
df_final['Additional_Info'].unique()
```

```python
df_final['Duration_Hour']=df_final['Duration'].str.split(' ').str[0].str.split('h').str[0]
df_final['Duration_Min']=df_final['Duration'].str.split(' ').str[1].str.split('m').str[0]
df_final.head()
```

```python
df_final[df_final['Duration']=='5m']
```

```python
df_final.drop(6474,axis=0,inplace=True)
df_final.drop(2660,axis=0,inplace=True)
```

```python
df_final['Duration_Min'].isnull().sum()
```

```python
df_final['Duration_Min']=df_final['Duration_Min'].fillna(0)
```

```python
df_final['Duration_Min'].isnull().sum()
```

```python
df_final['Duration_Hour']=df_final['Duration_Hour'].astype('int64')
df_final['Duration_Min']=df_final['Duration_Min'].astype('int64')
df_final.info()
```

```python
df_final['Duration_Hour']=df_final['Duration_Hour']+df_final['Duration_Min']/60
df_final.head()
```

```python
#Dropping the Duration and Duration_Min columns
df_final.drop('Duration',axis=1,inplace=True)
df_final.drop('Duration_Min',axis=1,inplace=True)
df_final.head()
```

```python
#Checkin the Airline column
df_final['Airline'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df_final['Airline']=labelencoder.fit_transform(df_final['Airline'])
df_final['Source']=labelencoder.fit_transform(df_final['Source'])
df_final['Destination']=labelencoder.fit_transform(df_final['Destination'])
df_final['Additional_Info']=labelencoder.fit_transform(df_final['Additional_Info'])
```
