### Missing Values

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
cat_features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype=='O']
for feature in cat_features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
## Replace missing value with a new label
def replace_cat_feature(df,cat_features_nan):
    cpy=df.copy()
    cpy[cat_features_nan]=cpy[cat_features_nan].fillna('Missing')
    return cpy

df=replace_cat_feature(df,cat_features_nan)

df[cat_features_nan].isnull().sum()
```

```python
df.head()
```

```python
## First lets handle numerical features which are missing
num_features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype!='O']

## We will print the numerical nan variables and percentage of missing values
for feature in num_features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values
for feature in num_features_nan:
    
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()
    
    ## create a new feature to capture nan values
    df[feature+'_nan']=np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[num_features_nan].isnull().sum()
```

```python
df.head()
```

```python
## Temporal Variables (Date Time Variables)
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    df[feature]=df['YrSold']-df[feature]
```

```python
df.head()
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

## Numerical Variables

Since the numerical variables are skewed we will perform log normal distribution

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature]=np.log(df[feature])
```

```python
df.head()
```

## Handling Rare Categorical Feature

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features
```

```python
for feature in categorical_features:
    temp=df.groupby(feature)['SalePrice'].count()/len(df)
    print(temp)
    temp_df=temp[temp>0.01].index
    print(temp_df)
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
df['MSZoning'].unique()
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```
